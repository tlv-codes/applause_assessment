```python
## search function ##
## prompts user for country, then device input, then returns testers, sorted by 'experience'

def find_testers():
    #
    print("What country are you interested in? (US,GB,JP,All?)\nSelect desired countries, separated by a comma.\n")
    country = input("Country: \n")
    clean_country = country.upper().split(',')
    if clean_country == ['ALL']:
        clean_country = ['US','JP','GB']
    tester_list = testers.loc[testers['country'].isin(clean_country)].reset_index()
    print("\nWhat device are you interested in? (iPhone 3,iPhone 4,iPhone 4S,iPhone 5,Galaxy S3,Galaxy S4,Nexus 4,Droid Razor,\nDroid DNA,HTC One) Enter desired devices, separated by a comma, or enter 'All'.")
    device = input("\nDevice: \n")
    clean_device = device.upper().split(',')
    if clean_device == ['ALL']:
        clean_device = ['IPHONE 4','IPHONE 4S','IPHONE 5','GALAXY S3','GALAXY S4','NEXUS 4','DROID RAZOR','DROID DNA','HTC ONE','IPHONE 3']
    devices_list = devices.loc[devices['description'].isin(clean_device)].reset_index()
    resulting = bugs.loc[(bugs['testerId'].isin(tester_list['testerId'])) & ((bugs['deviceId'].isin(devices_list['deviceId'])))]
    grouped = resulting.groupby(['deviceId','testerId']).count().sort_values(['deviceId','bugId'], ascending=[True,False])
    print('\n')
    print(grouped)
```

```python
# example 1: US testers, using iphone 3, 4, 5.
find_testers()
```

```python
# example 2: JP,GB testers using iPhone 4, Galaxy S3, and Nexus 4.
find_testers()
```

```python
# example 3: find testers for ALL countries, and ALL devices.
find_testers()
```
